```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Import our core library modules
from image_stitching.stitch import StitcherFactory

# create factory
stitcher_factory = StitcherFactory()
```

```python
# Initialize our stitcher
stitcher = stitcher_factory.createStitcher('planar')
stitchedImage = stitcher.stitch(['../imgs/1.jpg','../imgs/2.jpg','../imgs/3.jpg'])
plt.figure(figsize=(15, 10))
plt.imshow(cv2.cvtColor(np.clip(stitchedImage, 0, 255).astype(np.uint8), cv2.COLOR_BGR2RGB))
plt.title("Final Stitched Image (planar stitching)")
plt.axis('off')
plt.show()
```

```python
# Initialize our stitcher
stitcher = stitcher_factory.createStitcher('cylindrical')
stitchedImage = stitcher.stitch(['../imgs/1.jpg','../imgs/2.jpg','../imgs/3.jpg'])
plt.figure(figsize=(15, 10))
plt.imshow(cv2.cvtColor(np.clip(stitchedImage, 0, 255).astype(np.uint8), cv2.COLOR_BGR2RGB))
plt.title("Final Stitched Image (cylindrical stitching)")
plt.axis('off')
plt.show()
```
